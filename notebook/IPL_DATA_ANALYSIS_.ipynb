```python
plt.figure(figsize= (10,6))
sns.barplot(x = "Seasons", y = "count", hue = "toss_decision", data = season_toss_count_df)

## plotting barplot of "Season" vs "count" for bat & field both
```

<div class="alert alert-block alert-info"><p style='color:black;'> Most of the times, teams decide to feild first except in season 2009,2010,2013 where teams decided to bat first mostly.<br>

</div>

### 5) Number of times team has won the tournament :

```python
match_data.columns
```

```python
match_data["Seasons"].unique()
```

```python
df_2010 = match_data[match_data["Seasons"] == 2010]  # lets extract data of "2010" year only..

df_2010["winner"].tail(1).values[0]
```

```python
# dictionary to store winners of every season
winners_team = {}

for year in sorted(match_data["Seasons"].unique()):
    current_year_df = match_data[match_data["Seasons"] == year]
    winners_team[year] = current_year_df["winner"].tail(1).values[0]
```

```python
winners_team
```

```python
winners_team.values()
```

## 6) Most number of wins :

```python
match_data.columns
```

```python
match_data[["team1", "team2"]]
```

```python
match_data["team1"].value_counts()
```

```python
match_data["team2"].value_counts()
```

```python
matches_played = match_data["team1"].value_counts() + match_data["team2"].value_counts()

matches_played
```

```python
matches_played_df = matches_played.reset_index()
```

```python
matches_played_df.columns = ["team_name", "Matches_played"]
```

```python
matches_played_df
```

```python
wins = pd.DataFrame(match_data["winner"].value_counts()).reset_index()

wins.columns = ["team_name", "Wins"]

wins
```

```python
team_wins = matches_played_df.merge(wins, how = "inner", left_on= "team_name", right_on = "team_name")

team_wins
```

```python
# Bar-chart of above values
```

```python
trace1 = go.Bar(
    x = team_wins["team_name"],
    y = team_wins["Matches_played"],
    name = "Total Matches"
)

trace2 = go.Bar(
    x = team_wins["team_name"],
    y = team_wins["Wins"],
    name = "Matches Won"
)
```

```python
data = [trace1, trace2]
```

```python
iplot(data)
```
